```r
source("TMB_Funcs.txt")
# My Stuff!

path_name <- "/Volumes/CCG8/processed_data"
genome <- BSgenome.Hsapiens.UCSC.hg19
ensembl = useMart("ensembl", dataset = "hsapiens_gene_ensembl")
ensembl37 = useMart(host = "grch37.ensembl.org", biomart = "ENSEMBL_MART_ENSEMBL", dataset = "hsapiens_gene_ensembl")
select <- dplyr::select
```

### Loading Data (Which we think is WGS)

```r
data <- cbio_getData(path_name, "sclc", test = FALSE) %>% 
    filter(study == "sclc_ucologne_2015") %>% 
    select(-study)
```

```r
n <- 1000
gl_check <- full_join((data %>% 
    mutate(cr_value = paste0(Chromosome, ":", Start_Position, ":", End_Position)))[1:n,],
         (data %>% 
    pull(Hugo_Symbol))[1:n] %>% 
    {getBM(c("hgnc_symbol", "chromosome_name", "start_position", "end_position"), filters = c("hgnc_symbol"), values = ., mart = ensembl37)} %>% 
    filter(chromosome_name %in% c(1:22, "X", "Y")) %>% 
    mutate(gene_start = start_position, gene_end = end_position, Hugo_Symbol = hgnc_symbol) %>% 
    select(-c(start_position, end_position, hgnc_symbol)),
    by = "Hugo_Symbol") %>% 
    mutate(in_stated_gene = Start_Position >= gene_start & End_Position <= gene_end) %>% 
    select(Hugo_Symbol, Chromosome, Start_Position, in_stated_gene, End_Position, gene_start, gene_end, chromosome_name, Tumor_Sample_Barcode, cr_value) %>% 
    distinct() %>% 
    mutate(biomart_gene = map(cr_value, ~getBM("hgnc_symbol", "chromosomal_region", .x, ensembl37))) %>% 
    select(Hugo_Symbol, in_stated_gene, biomart_gene)
```
