```r
rll_sc_lung <- RLL_slide(maf_file = sc_lung_data[["training_data"]], starting_window = 2**20, iterations = 11, zoom = 2, coverage = 1)
```

```r
rll_sc_lung_aps <- RLL_analyse_panel_size(rll_sc_lung, sc_lung_data, panel_sizes = seq(from = 2**20, to = 5*(2**19), by = 2**19), max_windows = 10)
```
