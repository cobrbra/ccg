## Alternative Mechanisms of resistance development

For use in powerpoint for lunchtime seminar 13th November. Do tumours with TMB-High status develop resistance as a competing TMB-L subclone overtakes them, or because they incorporate some new mutation allowing escape from blockade?

```r
x <- data.frame(y = c(exp(-seq(0,4,0.02)**2),exp(seq(-1,1,0.01)**3)/4), z = c(rep("TMB-H Subclone",201),rep("TMB-L Subclone",201)), x = 1:201)
```

```r
ggplot(x, aes(x=x,y=y, fill = z))+ geom_area(position = "stack") + theme(axis.text.x = element_blank(), axis.text.y = element_blank())  + 
    theme(panel.background = element_rect(fill = 'grey18'), panel.grid.major = element_line(colour = "grey18"))+
    theme(axis.title.x = element_text(colour = "grey60", size = 30), axis.title.y = element_text(colour = "grey60", size = 30)) + labs(x = "Time", y = "Prevalence") +
    theme(plot.background = element_rect(fill = 'grey18'))
```

```r
x1 <- data.frame(y = c(exp(-seq(0,4,0.02)**2)/2,rep(0,50),exp(seq(-3,0,0.02)**3/2)/2), z = c(rep("TMB-H Subclone",201),rep("TMB-H + Mut",201)), x = 1:201)
```

```r
ggplot(x1, aes(x=x,y=y, fill = z))+ geom_area(position = "stack") + theme(axis.text.x = element_blank(), axis.text.y = element_blank())  + 
    theme(panel.background = element_rect(fill = 'grey18'), panel.grid.major = element_line(colour = "grey18"))+
    theme(axis.title.x = element_text(colour = "grey60", size = 30), axis.title.y = element_text(colour = "grey60", size = 30)) + labs(x = "Time", y = "Prevalence") +
    theme(plot.background = element_rect(fill = 'grey18')) + scale_y_sqrt()
```

## Visualising a sliding window algorithm

```r
x <- tibble(x = runif(1000), y = rexp(1000))
```

```r
ggplot(x,aes(x=x,y=y)) + geom_point(colour = "red") + ylim(c(0,20)) + theme_bw() + theme(panel.border = element_blank(), panel.grid.major = element_blank(),
panel.grid.minor = element_blank(), axis.line = element_line(colour = "black"))
```

```r
X <- tibble(x = 5*rexp(100), y = 0.5*x + rnorm(100))
X <- X %>% 
    filter(y>0)
```
